## Stack

```python
#  creating a multilevel dataframe
header = pd.MultiIndex.from_product([['Sem1','Sem2'],['3_Months','6_Months'],['ML','Python']])
d = [[12,45,67,56,45,34,23,67],[78,89,45,67,45,23,78,57],[45,67,89,90,24,57,46,97],[67,44,56,55,13,43,62,84]]
df = pd.DataFrame(d,index = 'James Anil Sunil Smriti'.split(),columns = header)
df
```

```python
df.columns
```

```python
df.index
```

```python
df1 = df.stack()
```

```python
df1.loc[('Sunil','ML'),('Sem1','6_Months')]
```

```python
stacked_df1 = df.stack(level=1)
stacked_df1
```

```python
stacked_df1.loc[('James','6_Months'),('Sem1','Python')]
```

```python
un_stacked = stacked_df1.unstack(level=1)
un_stacked
```

## Melt

```python
data = {
    "location":['Bangalore','Chennai'],
    "Temperature":['Prdict','Actual'],
    "Jan-2019":[30,42],
    "Feb-2019":[45,43],
    "Mar-2019":[24,22],
    }
        
df = pd.DataFrame(data)
df
```

```python
df2 = pd.melt(df,id_vars = ['location','Temperature'],var_name="Data1",value_name = 'Value1')
df2
```

```python
df3 = pd.pivot_table(df2, index = ['location','Temperature'],aggfunc = 'mean')
df3
```

```python
df4 = pd.pivot_table(df2, index = ['location','Temperature'],aggfunc = 'sum')
df4
```

```python
d={
   'name':['james','anil','sunil','anil'],
   'exam-months':['3_months','3-months','6_months','6_months'],
   'subjects':['ml','python','sql','python'],
   'score':[62,47,55,74],
   'score1':[162,147,155,174]
   
}

df6=pd.DataFrame(d)
print(df6)
```

```python
df7 = pd.pivot_table(df6,index=['name','subjects'],aggfunc='mean')
df7
```

```python
df7.loc[('james','ml'),'score1']
```

```python
df8 = df6.pivot_table(index = 'name',columns = 'subjects',values = 'score',aggfunc = 'mean')
df8
```
