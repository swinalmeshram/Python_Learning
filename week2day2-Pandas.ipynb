```python
import pandas as pd
df6 = pd.read_csv('profit.csv')
df6
```

```python
df6['rank'] = df6['Tech1'].rank(ascending=0)
df6
```

```python
df6['minrank'] = df6['Tech1'].rank(ascending=0,method='min')
df6
```

```python
df6['maxrank'] = df6['Tech1'].rank(ascending=0,method='max')
df6
```

```python
df6['denserank'] = df6['Tech1'].rank(ascending=0,method='dense')
df6
```

```python
df6['avgrank'] = df6['Tech1'].rank(ascending=0,method='average')
df6
```

```python
df6['%rankrank'] = df6['Tech1'].rank(ascending=0,method='average',pct=True)
df6
```
