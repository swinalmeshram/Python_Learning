## Line Plot


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
x1 = np.array([2,4,6,8,10])
y1 = x1*2
x2 = x1
y2 = x1**2

plt.plot(x1,y1,linestyle = 'solid',label = 'plot1',marker='+',color = 'r')
plt.plot(x2,y2,linestyle='dotted',label = 'plot2',marker = '>',color = 'g')
plt.xlabel('X-axis')
plt.ylabel('Y-axxis')

plt.title('First Plot')
plt.legend(fontsize = 18,loc=0)
plt.savefig('myplot.png')
plt.show()
print()
```

```python
plt.scatter(y1,x1,label = 'temp',marker='+',color = 'r')
# plt.plot(x2,y2,linestyle='dotted',label = 'plot2',marker = '>',color = 'g')
plt.xlabel('X-axis')
plt.ylabel('Y-axxis')

plt.title('Scatter Plot')
plt.legend(fontsize = 18,loc=0)

plt.show()
print()
```

```python
stock_names = ['goog', 'appl', 'IBM', 'INTC', 'CSCo']
stock_value = [1000 ,800 ,600, 400, 200]
x_pos = range(len(stock_names))


plt.bar(x_pos,stock_value,width = 0.35,align = 'center',color = 'b',alpha = 0.8)
plt.xticks(x_pos,stock_names)
plt.ylabel('price')
plt.xlabel('Name')
plt.title('Stock Prices')
plt.show()
```

```python
vc = df['SaleCondition'].value_counts()
print(vc)
vc.plot(kind = 'bar',colors = 'r')
plt.title('Bar plot House Train Small')
plt.xlabel('Salecondtion')
plt.ylabel('No. of sales')
```

```python
vc = df[['SaleCondition','BldgType']].apply(pd.value_counts)
print(vc)
vc.plot(kind = 'bar',colors = 'rg')
plt.title('Bar plot House Train Small')
plt.xlabel('Salecondtion')
plt.ylabel('No. of sales')
```

```python
print(df['BldgType'])
vc = df['BldgType'].value_counts()
print(vc)
explodelist = [0] * len(vc)
explodelist[1] = 0.5
print(explodelist)
vc.plot(kind = 'pie',explode = explodelist,autopct = '%1.1f',labels=vc)
plt.title('Types of buildings')
```
